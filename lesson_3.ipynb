```python
horizontal = [1, 1, -1, -1, 2, 2, -2, -2]
vertical = [2, -2, 2, -2, 1, -1, 1, -1]

class Knight:
    def __init__(self, x = 0, y = 0, n = 8):
        self.cx = x
        self.cy = y
        self.n = n
        
        self.access = [[0] * n for _ in range(n)]
        self.board = [[0] * n for _ in range(n)]
        
        # Initialize access matrix
        for i in range(n):
            for j in range(n):
                for a, b in zip(horizontal, vertical):
                    if self.in_board(i+a, j+b):
                        self.access[i][j] += 1
        
        print("The starting access matrix: ")           
        self.print(self.access)
    
    def in_board(self, x, y):
        return x >= 0 and x < self.n and y >= 0 and y < self.n
        
    def print(self, m):
        n = len(m)
        
        for i in range(n):
            for j in range(n):
                print(f'{m[i][j]: <5}', end='')
            print()
    
    def get_tour(self):
        
        step = 1
        x, y = self.cx, self.cy
        self.board[x][y] = step
        
        while step < self.n ** 2:
            min_access, min_x, min_y = 9, 0, 0
            
            for a, b in zip(horizontal, vertical):
                tx = x + a
                ty = y + b
                
                if self.in_board(tx, ty) and self.board[tx][ty] == 0:
                    self.access[tx][ty] -= 1
                    
                    if self.access[tx][ty] < min_access:
                        min_access = self.access[tx][ty]
                        min_x, min_y = tx, ty
            step += 1
            
            x, y = min_x, min_y
            self.board[x][y] = step
        
        print("Solution found: ")
        self.print(self.board)
```

```python
my_knight = Knight(1, 1, 5)
my_knight.get_tour()
```
